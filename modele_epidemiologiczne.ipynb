# O epidemiach 
<p style = "text-indent: 20px;">  Przypadki epidemii, czyli występowania u ludzi zachorowań na określoną chorobę w określonym czasie i na określonym terenie w liczbie przypadków większej niż przeciętnie$^{[1]}$, zostały odnotowane już w starożytności i miały znaczący wpływ na rozwój ludzkości. </p>

# Podstawy teoretyczne 

## Analityczne rozwiązywanie równań różniczkowych:

### Metoda czynnika całkującego 

Pozwala znaleźć rozwiązania niektórych równań różniczkowych pierwszego rzędu. <br> 
Weźmy rówanie $y' + P(t)y = Q(t)$. <br>
Schemat działania: <br>
<li> Znajdujemy $\mu(t) = exp(\int P(t)dt) $ </li>
<li> Mnożymy równanie liniowe przez $ \mu (t) $ </li>
<li> Przekształamy równanie do $ ( \mu (t)y)' = \mu P(t)Q(t) $ </li>
<li> Całkujemy obie strony równania i otrzymujemy: $ y(t) = \mu (t)^{-1} \int \mu (t)Q(t)dt $ </li>


### Równanie różniczkowe Bernoulliego 

Równanie różniczkowe zwyczajne postaci: $y' + P(x)y = Q(x)y^n $ , gdzie $n$ jest liczbą rzeczywistą różną od $0$ i $1$.<br>
Schemat działania: <br>
<li> Dzielimy obie strony równiania przez $y^n$ i otrzymujemy: $ \frac {1}{y^n} y' + P(x)y^{1-n} + Q(x) = 0 $ </li>
<li> Wprowadzamy pomocniczą zmienną zależną $z = y^{1-n} $. Wówczas: $z'= (1-n)y^{-n}y'$ </li>
<li> Po wstawieniu tej zmiennej i jej pochodnej do równania Bernoulliego otrzymujemy: $\frac{1}{1-n} z' + P(x)z + Q(x) = 0 $. </li>
<li> Otrzymujemy równianie różniczkowe liniowe niejednorodne i jako takie je rozwiązujemy. </li>


## Numeryczne rozwiązywanie równań różniczkowych:

### Metoda Rungego-Kutty  

Metoda numerycznego rozwiązywania nieliniowych równań różniczkowych charakteryzująca się tym, że przy obliczaniu wartości funkcji w kolejnym punkcie $x_i = x_0 + ih$ bierze się pod uwagę również punkty wewnątrz $i$-tego kroku. W zależności od tego, ile tych punktów bierze się pod uwagę, mówi się o rzędzie metody.


### Metoda RK 4.rzędu

Dane jest równanie różniczkowe pierwszego rzędu: $ y'=f(x,y),$ z wartościami początkowymi $ y(x_{0})=y_{0} $. 

Wybieramy krok całkowania h i wprowadzamy następujące oznaczenia: $x_i = x_0 + ih$, $y_i = y(x_i), i=0,1,2,...$

Zgodnie z metodą kolejne wartości $y_i$ szukanej krzywej całkowej $y = y(x)$ znajdujemy ze wzoru:

$$ y_{n+1}=y_{n}+ \Delta y_{n},$$
gdzie:
$$ \Delta y_{n}=\frac {1}{6}(k_{1}+2k_{2}+2k_{3}+k_{4}),$$
oraz:

$ k_{1}=hf\left(x_{n},y_{n}\right),$

$ k_{2}=hf\left(x_{n}+{\frac {h}{2}},y_{n}+{\frac {1}{2}}k_{1}\right),$

$ k_{3}=hf\left(x_{n}+{\frac {h}{2}},y_{n}+{\frac {1}{2}}k_{2}\right),$

$ k_{4}=hf\left(x_{n}+h,y_{n}+k_{3}\right).$

Jak widać, wartość $y_{n+1}$ zależy od wartości $y_n$ i $h$ $^{[6]}$. 


### Metoda Eulera - metoda RK 1. rzędu 
    
Metoda Eulera jest szczególnym przypadkiem metod Runge-Kutty (metoda Eulera pojawiła się jako pierwsza, dlatego zachowano tradycyjną nazwę) w którym kolejne wartości $y_i$ znajdujemy ze wzoru:


$$ y_{n+1}=y_{n}+ \Delta y_{n}.$$
gdzie:
$$ \Delta y_{n}= hf\left(x_{n},y_{n}\right) $$.

# Podział modeli epidemiologicznych

Modele epidemiologiczne pozwalają na przewidywanie rozwoju epidemii. Umożliwiają one określenie czasu jej trwania, szybkości jej rozwoju czy też aktualnej liczby osobników zarażonych lub zdrowych. 

Wyróżniamy kilka modeli epidemiologicznych, podzielonych według ilości grup na które została podzielona populacja:
- SI SIS
- SIR SIRS
- SEIR SEIRS
    
Celem modelowania jest opisanie ewolucji populacji w każdej z grup. 


Każdy model możemy rozpatrywać na dwa sposoby:

- Bez dynamiki życiowej, używany w przypadkach gdy populacja jest niewielka, czas trwania epidemii jest najprawdopodobniej krótki, a epidemia mało śmiertelna.
- Z dynamiką życiową, gdy epidemia jest śmiertelna, jej czas trwania będzie długi, więc możliwe, że i wielopokoleniowy.

# Model SI oraz SIS 

## Model SI 

Model SI to najprostszy model epidemiologiczny (pasuje do takich chorób jak wirus cytomegalii (CMV) lub opryszczki). W modelach SI oraz SIS populacja podzielona jest na dwie grupy:
- S - osobników zdrowych i podatnych (ang. susceptible) na infekcję,
- I - osobników zainfekowanych (ang. infected) <br>

#### Oznaczenia:

- $S(t)$ - liczba osobników zdrowych (podatnych) w czasie t;
- $I(t)$ - liczba osobników zainfekowanych w czasie t;
- $N(t)$ - liczebność populacji $N(t)=S(t)+I(t);$
- $\beta>0$ - współczynnik przenoszenia choroby.
- $\nu>0$ - współczynnik rozrodczości i śmiertelności (w przypadku modelu z dynamiką życiową);    

#### Założenia:

Liczba populacji jest stała. Choroba rozprzestrzenia się w zamkniętej populacji, której liczebność jest stała, a okres inkubacji jest na tyle krótki, że można go zaniedbać tzn. osobnik podatny, który się zaraził, natychmiast zaczyna chorować (przechodzi do grupy $I$). Gdy zainfekowany zetknie się z podatnym, może wystąpić przeniesienie choroby. Funkcja $f(S, I)$, która opisuje ten proces, nazywa się funkcją zapadalności (zachorowalności). Zapadalność jest zdefiniowana jako liczba nowo zarejestrowanych przypadków danej choroby w przedziale czasu, więc f jest współczynnikiem. Składa się z dwóch elementów: 
 - liczby zachodzących kontaktów,
 - opisu prawdopodobieństwa, że taki kontakt spowoduje przeniesienie choroby. 
    
Znalezienie funkcji, która odpowiednio opisuje proces kontaktu, nie jest łatwe. Jedna z popularnych przybiera formę

$$f(S,I)=\beta S \label{eq1}\tag{1}I$$

Założeniem tego modelu jest to, że każdy zainfekowany może spotkać każdego podatnego. W wystarczająco dużej populacji jest to nierealne (dla ludzi lub zwierząt), a inna stosowana funkcja ma postać 

$$f(S,I)=\beta \frac{S I}{N} $$

Oznacza to, że każdy zainfekowany może spotkać tylko część podatnych  $^{[7]}$.

Dla uproszczenia obliczeń, w opisywanych poniżej modelach zastosowałyśmy formę ($1$). 
   
#### Dodatkowe założenia w przypadku modelu z dynamiką życiową:

- osobniki nowonarodzone są podatne ($S$);
- liczebność populacji jest stała i wynosi $N$ (rodzi się tyle samo osobników co umiera);
- podatne osobniki rodzą się w tempie $\nu$ proporcjonalnym do całkowitej populacji $N$, umierają w tempie $\nu$ proporcjonalnym do podatnej populacji $S$;
- zainfekowani giną w tempie $\nu$, proporcjonalnym do populacji zainfekowanej $I$ oraz nigdy nie wyzdrowieją z choroby. 

Biorąc pod uwagę wszystkie te hipotezy i wybierając funkcję zachorowalności, model ma schemat przepływu pokazany na poniższym rysunku. Możemy łatwo uzyskać układ równań z tego schematu, pisząc, że dla konkretnej grupy ($S$ lub $I$) strzałki przychodzące oznaczają zmianę z plusem, a strzałki wychodzące - z minusem.

Teraz z łatwością możemy napisać układ równań różniczkowych dla wybranego przez nas modelu <b>SI z dynamiką życiową</b>, którym będziemy się dalej zajmować

$$
\begin{align}
&\frac{dS}{dt}=\nu N - \beta S I - \nu S \label{2a}\tag{2a} \\
&\frac{dI}{dt}=\beta S I - \nu I. \label{2b}\tag{2b}
\end{align}
$$

<h2> Porównanie modelu SI z modelem SIS </h2>

Z powyższego rysunku możemy łatwo zauważyć podobieństwo pomiędzy oboma modelami. Dodatkowym założeniem dla modelu SIS jest sytuacja, w której zainfekowana osoba może wyzdrowieć, czyli przejść z grupy $I$ do $S$. Napiszmy teraz układ równań różniczkowych dla modelu SIS bez dynamiki życiowej 

$$
\begin{align}
&\frac{dS}{dt}= - \beta S I + \xi I \\
&\frac{dI}{dt}=\beta S I - \xi I
\end{align}
$$

Równanie $(2a)$ możemy zapisać jako

$$ \frac{dS}{dt}=\nu N - \beta S I - \nu S = - \beta S I +\nu(N-S) = - \beta S I + \nu I $$

Widzimy więc, że powyższe równania, odpowiednio dla $S$ oraz $I$ są analogiczne, różnią się tylko znaczeniem współczynnika oraz interpretacją. Sytuacja dla modelu SIS z dynamiką życiową jest już nieco inna.

## Rozwiązanie analityczne modelu SI z dynamiką życiową

### Całkowita populacja jest stała

W wybranym przez nas modelu przyjmujemy warunki początkowe $S(0) = S_0 \ge 0$ oraz $I(0) = I_0 \ge 0.$
Rozważany układ $(2)$ jest układem nieliniowym. Zwykle analizuje się takie układy używając metod jakościowych, jednak w tym przypadku można znaleźć dokładne rozwiązanie. Na początek pokażemy, że całkowita populacja jest stała, mamy (dla uposzczenia $ ' = \frac{d}{dt}$):

\begin{equation}
\begin{split}
N'
& = S' + I' \\
& = \nu N - \beta S I - \nu S + \beta S I- \nu I\\
& = \nu N - \nu (S+I)\\
& =0
\end{split}
\end{equation}

Czyli dla każdego $t, N(t)\equiv N_0 = S_0+I_0$.

<br> 

### Równanie dla I - równanie Bernoulliego

Wiedząc, że $S+I=N$ możemy przekształcić równanie $(2b)$ do następującej postaci

$$I'-(\beta N - \nu)I=-\beta I^2$$

Jest to równanie różniczkowe Bernouliiego. Podstawiamy, więc $u=I^{-1}$ wtedy $u'=-I^{-2}I'$. Dzieląc powyższe równanie przez $I^2$ oraz podstawiając $u$ i $u'$ otrzymujemy równanie różniczkowe liniowe

$$u'+(\beta N-\nu)u=\beta. \label{3}\tag{3}$$

<br>

### Rozwiązywanie równania różniczkowego liniowego rzędu pierszego metodą czynnika całkującego

Równanie $(3)$ jest postaci $$x'+P(t)x=Q(t)$$

gdzie $P(t)=\beta N-\nu$ oraz $Q(t)=\beta$. Czynnik całkujący wynosi wtedy

 $$\mu(t)=exp\left(\int P(t) dt\right)=e^{(\beta N-\nu)t}$$
 
Mnożąc obie strony równanie $(3)$ przez $\mu(t)$ otrzymujemy

$$ (\mu(t)u)'=\beta e^{(\beta N-\nu)t} $$

<li> <b> Pierwszy przypadek gdy $\beta N-\nu \neq 0$ </b>

Całkując stronami

$$\mu(t)u=\frac{\beta}{\beta N-\nu} e^{(\beta N-\nu)t} +C,\hspace{20pt} gdzie  \hspace{10pt} C \in \mathbb{R}$$

wtedy
$$u=\frac{\beta}{\beta N-\nu} + C e^{-(\beta N-\nu)t} $$

Tutaj łatwiej jest znaleźć wartość $C$ przed napisaniem rozwiązania względem $I$. Wiedząc, że $u(0)=1/I(0)=1/I_0$, wtedy

$$u(0)=\frac{1}{I_0}=\frac{\beta}{\beta N-\nu}+C$$

co implikuje 

$$C=\frac{\beta N-\nu-\beta I_0}{I_0 (\beta N-\nu)}$$

Zatem rozwiązanie równania $(3)$ jest zadane nastepująco

\begin{equation}
\begin{split}
u
& = \frac{I_0\beta+((\beta N-\nu)-I_0\beta)e^{-(\beta N-\nu)t}}{I_0 (\beta N-\nu)} \\
& = \frac{I_0\beta(1-e^{-(\beta N-\nu)t})+(\beta N-\nu)e^{-(\beta N-\nu)t}}{I_0 (\beta N-\nu)} \\
\end{split}
\end{equation}

Rozwiązanie początkowego równaia ma postać ($I=1/u$)

$$I(t)=\frac{I_0 (\beta N-\nu)}{I_0\beta(1-e^{-(\beta N-\nu)t})+(\beta N-\nu)e^{-(\beta N-\nu)t}}  $$

<li> <b> Drugi przypadek gdy $\beta N-\nu = 0$ </b>

Wtedy równanie $(3)$ przyjmuje postać

$$u'=\beta,$$

którego rozwiązanie jest bardzo proste

$$u=\beta t+C.$$

Po obliczeniu $C$ oraz podtawieniu $I=\frac{1}{u}$ otrzymujemy rozwiązanie

$$I(t)=\frac{I_0}{\beta I_0 t+1} $$

<br><br>

Podsumowując, rozwiązanie układu ($2$) wygląda następująco
- gdy $\beta N-\nu \neq 0$

$$S(t)=N-\frac{I_0 (\beta N-\nu)}{I_0\beta(1-e^{-(\beta N-\nu)t})+(\beta N-\nu)e^{-(\beta N-\nu)t}} $$
$$I(t)=\frac{I_0 (\beta N-\nu)}{I_0\beta(1-e^{-(\beta N-\nu)t})+(\beta N-\nu)e^{-(\beta N-\nu)t}}  $$

- gdy $\beta N-\nu = 0$

$$S(t)=N-\frac{I_0}{\beta I_0 t+1} $$
$$I(t)=\frac{I_0}{\beta I_0 t+1} $$

## Analiza modelu SI z dynamiką życiową

### Bazowy współczynnik reprodukcji 

Obserwując powyższe równania, zauważamy dwa przypadki:
- jeśli $\beta N-\nu \leq 0$, wtedy ułamki dążą do $0$, ponieważ gdy $\beta N - \nu < 0 $  to $\lim_{t \to \infty} e^{-(\beta N-\nu)t}=+\infty$ wtedy w mianownikach otrzymujemy $-\infty$, a gdy $\beta N-\nu = 0$ to w mianowniku otrzymujemy $+\infty$ przy $t \rightarrow +\infty$, w takim wypadku:

$$\lim_{t \to \infty} S(t)=N, \hspace{30pt} \lim_{t \to \infty} I(t)=0$$ 

- jeśli $\beta N-\nu>0$, to $\lim_{t \to \infty} e^{-(\beta N-\nu)t}=0$ więc 

$$\lim_{t \to \infty} S(t)=\frac{\nu}{\beta}, \hspace{30pt} \lim_{t \to \infty} I(t)=N-\frac{\nu}{\beta}$$ 

Możemy przeformułować ten wniosek używając ważnego w epidemiologii współczynnika reprodukcji, zwykle oznaczanego $R_0$. W zależności czy uwzględniamy dynamiką życiową czy nie, współczynnik będzie nieco się różnił. W omawianym modelu z dynamiką życiową przybiera formę

$$R_0=\frac{\beta N}{\nu}$$ 

wtedy

$$R_0 \leq 1 \Leftrightarrow	\beta N-\nu \le 0$$
$$R_0>1 \Leftrightarrow	\beta N-\nu > 0.$$

Za pomocą współczynnika $R_0$ możemy określić zachowanie się epidemii:

- jeśli $R_0 \leq 1$, to epidemia się wygasza, ponieważ liczba zainfekowanych po pewnym czasie spada do zera

$$\lim_{t \to \infty} S(t)=N, \hspace{30pt} \lim_{t \to \infty} I(t)=0$$ 

- jeśli $R_0>1$, to epidemia staje się endemię, ponieważ liczba zainfekowanych po pewnym czasie pozostaje na stałym poziomie 

$$\lim_{t \to \infty} S(t)=\frac{N}{R_0}, \hspace{30pt} \lim_{t \to \infty} I(t)=N-\frac{N}{R_0}.$$ 

Bazowy współczynnik reprodukcji $R_0$ określa zatem, czy choroba ustabilizuje się w populacji, czy też nie.
Innymi słowy, $R_0$ jest średnią liczbą wtórnych infekcji powstających, gdy osoba zarażona wchodzi do populacji podatnych ludzi (czyli ile osób jedna zarażona osoba może zarazić). W przypadku, gdy  $R_0>1$, możemy zauważyć, że im większy $R_0$, tym wyższy odsetek zakażeń w populacji. Zatem, poza tym, że jest on progiem inwazji choroby, jest również wskaźnikiem zaraźliwość choroby $^{[7]}$.

# Model SIR oraz SIRS  

## Model SIR

Model SIR jest kolejnym modelem epidemiologicznym. Został on po raz pierwszy zastosowany przez Kermacka i McKendricka w 1927 roku, a następnie został użyty do opisu rozwoju wielu chorób, takich jak odra, świnka czy różyczka, czyli chorób dziecięcych przenoszonych drogą powietrzną $^{[8]}$. Można zauważyć, że na choroby te można zachorować tylko raz - po wyzdrowieniu nabieramy odporoność na całe życie. Łatwo dostrzec więc różnicę pomiędzy wcześniej opisywanym modelem SI - w modelach SIR oraz SIRS choroba jest uleczalna, a  populacja podzielona jest na trzy grupy:
- S - osobników zdrowych i podatnych (ang. susceptible) na infekcję,
- I - osobników zainfekowanych (ang. infected),
- R - osobników wyleczonych i odpornych (ang. recovered). <br> 

<img src="scheme/SIR.PNG"> 

Równania modelu bez dynamiki życiowej bazującego na założeniach analogicznych do modelu SI mają postać: 

$$
\begin{align}
&\frac{dS}{dt} = - \beta S I  \label{eq4a}\tag{4a} \\
&\frac{dI}{dt} = \beta S I - \gamma I   \label{eq4b}\tag{4b} \\
&\frac{dR}{dt} = \gamma I  \label{eq4c}\tag{4c}
\end{align}
$$

gdzie:
- $N$ - liczebność populacji, czyli $S+I+R$
- $\beta$ - współczynnik (tempo) rozprzestzeniania się infekcji ($ \beta > 0$)
- $\gamma$ - współczninnik zdrowienia osobników zainfekowanych.

<br>

## Analiza układu SIR bez dynamiki życiowej
### Bazowy współczynnik reprodukcji 

Rozmawialiśmy o nim przy modelu SI, jednak w tym modelu będzie on wyglądał nieco inaczej. Nie jesteśmy w stanie podać prostego analitycznego rozwiązania rozważanegu układu $(4)$, jednak możemy przeanalizować poszczególne równania różniczkowe. Chcemy się dowiedzieć czy przy zadanych paramatrach $\beta$ i $\gamma$ oraz wartościach początkowych $S_0$ i $I_0$, infekcja rozprzestrzeni się, czy też nie. A jeśli rozprzestrzeni, to jak będzie przebiegał jej rozwój w czasie i kiedy zacznie zanikać. Na to pytanie może odpowiedzieć nam zmiana ilości osób zainfekowanych, czyli $I'$. Jeśli będzie ona większa od zera to znaczy, że liczba chorych rośnie, jeśli mniejsza, liczba chorych maleje. Z równania $(4b)$ wynika, że:

$$\left.\frac{dI}{dt}\right|_{t=0}=I_0(\beta S_0 -\gamma)\begin{cases}>0,\\<0,\end{cases} \hspace{20pt} \textrm{jeśli } S_0 \begin{cases}>\rho,\\<\rho,\end{cases} \label{5}\tag{5} $$

gdzie $\rho=\gamma/\beta$. Zauważmy, że prawa strona równania $(4a)$ jest zawsze mniejsza od zera, ponieważ liczba osób w każdej grupie jest zawsze nieujemna oraz współczynnik $\beta$ jest dodatni, oznacza to, że liczba osób podatnych zawsze będzie maleć, czyli będzie mniejsza od $S_0$, czyli początkowej liczby osób podatnych. Można zauważyć, że nie jest to prawdą przy omawianym wcześniej układzie SI z dynamiką życiową (tam $ S'=\nu N - \beta S I - \nu S = - \beta S I + \nu I=$ $=I(\nu -\beta S)$ mogło przyjmować wartości różnych znaków). Jeśli zatem $S_0$ będzie mniejsze od $\rho$ to, każde kolejne $S$ również będzie od $\rho$, a wtedy liczba osób zainfekowanych $I$ będzie ciągle maleć, zapisujemy:

$$\textrm{jeśli} \hspace{5pt} S_0\leq \frac{\gamma}{\beta} \hspace{5pt}\textrm{to}\hspace{5pt} \frac{dI}{dt}=I(\beta S -\gamma) \leq 0\hspace{10pt} \textrm{dla wszystkich  } t\geq0.$$

Wtedy każde kolejne $I$ będzie mniejsze od poprzedniego (oraz nie mniejsze niż $0$), zapisujemy $I_0>I\rightarrow0$ przy $t\rightarrow\infty$, a zatem infekcja zanika, czyli epidemia się nie pojawia. Natomiast jeśli $S_0>\rho$ to $I$ początkowo rośnie i wtedy wybucha epidemia. Termin epidemia oznacza, że $I>I_0$ dla pewnego $t>0$. Ilość osób zainfekowanych będzie rosła do wartości maksymalnej, a następnie będzie maleć (o tym w dalszej części). Możemy więc stwierdzić, że istnieje zjawisko progowe. Jeśli $S_0>\gamma/\beta$, to mamy epidemię, a jeśli $S_0<\gamma/\beta$ to nie. Parametr krytyczny $\gamma/\beta$ nazywa się czasami względnym współczynnikiem zdowienia i stanowi odwrotność parametru $\sigma=\beta/\gamma$, czyli współczynnika kontaków. Mnożąc powyższe nierówności przez $\beta/\gamma$ nasuwa się pomysł aby oznaczyć

$$R_0=\frac{\beta S_0}{\gamma}$$

gdzie $R_0$ opisuje bazowy współczynnik reprodukcji dla danej infekcji, czyli liczbę osobników wtórnie zainfekowanych przypadającą na jednego osobnika pierwotnie zainfekowanego w całej populacji osobników podatnych. W powyższym wzorze parametr $1/\gamma$ odzwierciedla średni czas infekcji. Jeśli jedna infekcja pierwotna prowadzi do więcej niż jednej infekcji wtónej, co oznacza $R_0>1$ to choroba się rozprzestrzenia.

Porównajmy jak zmienia się liczba każdej z grup (podatni, zainfekowani, wyzdrowiali) w zależności od współczynnika reprodukcji. Na poniższych wykresach wygenerowanych przy pomocy metody Rungego-Kutty 4. rzędu, widać, że gdy  $R_0≤1$  epidemia się nie pojawia, w przeciwny wypadku epidemia wybucha, a liczba zainfekowanych jest tym większa im większe $R_0$ $^{[5]}$.
<br>


```python
#Porównanie zachowania liczby podatnych, zainfekowanych, i ozdrowiałych w zależności od wsp. reprodukcji w modelu SIR 

def S(x,s,i,r):
    return -beta*s*i                            
def I(x,s,i,r): 
    return beta*s*i - gamma*i
def R(x,s,i,r):
    return gamma*i

S_0, I_0, R_0 = 990, 10, 0

h = 0.1           
N_t = 30 #dni  

beta,gamma=0.0005,0.5
S2_0,I2_0=1000,100
czas, S1, I1, R1 = rk4_method_3(S,I,R,S2_0,I2_0,R_0)
fig, axes = plt.subplots(1,3, figsize=(16,5))
axes[0].plot(czas,S1,'g',czas,I1,'r',czas,R1,'b')
axes[0].set_xlabel('czas (dni)', fontsize=13)
axes[0].set_ylabel('liczba osobników',fontsize=13)
axes[0].legend(['S','I','R'],loc="right")
axes[0].set_title("$R_0=$"+str(round(beta*S2_0/gamma,2))+"      $beta = $"+str(round(beta,4))+"      $gamma = $"+str(round(gamma,4)),fontsize=14)

beta,gamma=0.001,0.4
czas, S1, I1, R1 = rk4_method_3(S,I,R,S_0,I_0,R_0)
axes[1].plot(czas,S1,'g',czas,I1,'r',czas,R1,'b')
axes[1].set_xlabel('czas (dni)',fontsize=13)
axes[1].set_ylabel('liczba osobników',fontsize=13)
axes[1].legend(['S','I','R'],loc="right")
axes[1].set_title("SIR bez dynamiki życiowej\n\n$R_0>1$\n\n"+"$R_0=$"+str(round(beta*S_0/gamma,2))+"      $beta = $"+str(round(beta,4))+"      $gamma = $"+str(round(gamma,4)),fontsize=14)

beta,gamma=0.001,0.2
czas, S1, I1, R1 = rk4_method_3(S,I,R,900,100,R_0)
axes[2].plot(czas,S1,'g',czas,I1,'r',czas,R1,'b')
axes[2].set_xlabel('czas (dni)',fontsize=13)
axes[2].set_ylabel('liczba osobników',fontsize=13)
axes[2].legend(['S','I','R'],loc="right")
axes[2].set_title("$R_0=$"+str(round(beta*S_0/gamma,2))+"      $beta = $"+str(round(beta,4))+"      $gamma = $"+str(round(gamma,4)),fontsize=14)

plt.tight_layout()

plt.show()
```

### Maksymalna liczba zainfekowanych

Powróćmy do sytuacji gdy $S_0>\rho$. Wiemy, że wtedy liczba chorych $I$ będzie początkowo rosła, jednak do jakiej liczby? Liczba chorych przestanie rosnąć, gdy ich zmiana będzie równa $0$, chcemy więc policzyć kiedy $I'=0$.  

$$I'=\beta S I - \gamma I =I(\beta S - \gamma) =0, \hspace{15pt} \textrm{więc} \hspace{15pt} S=\frac{\gamma}{\beta}=\rho, \hspace{15pt} \textrm{bo} \hspace{15pt} I\neq0$$

Dla $I=0$ otrzymujemy punkty krytyczne, wtedy $I$ już nigdy się nie zmieni i pozostanie $0$, (analiza tej sytuacji w dalszej części), dlatego w poszukiwaniu maksimum, takie $I$ nas nie interesuje. Znamy więc liczbę osób podatnych $S$ gdy liczba chorych jest maksymalna, ale wciąż nie wiemy ile ona wynosi. Zauważmy, że całkowita liczba populacji to $N=S+I+R$, więc nie możemy łatwo wywnioskować liczby chorych $I$, bo nie znamy również liczby osób wyzdrowiałych $R$. Z pomocą przychodzi ciekawy rezultat z początkowych równań $(4a)$ i $(4b)$ w naszym modelu, podzielmy je przez siebie:

$$\frac{dI}{dS}=-\frac{(\beta S-\gamma)I}{\beta S I}=-1+\frac{\rho}{S}, \hspace{15pt} \rho=\frac{\gamma}{\beta}, \hspace{12pt}(I\neq0). $$

Całkując powyższe równanie, otrzymujemy orbity w przestrzeni fazowej $(S,I)$ opisane zależnością

$$I+S-\rho ln S=const=I_0+S_0-\rho ln S_0,   \label{6}\tag{6} $$

z wykorzystaniem warunków początkowych. Orbity te wygenerowano poniżej. Zauważmy, że $I_0+S_0=N$, ponieważ $R_0=0$. W powyższym równaniu występują tylko dwie zmienne zależne z czego dla szukanej sytuacji $S$ jest znana i wynosi $\rho$, możemy teraz wyznaczyć maksymalną wartość $I$ 

\begin{equation}
\begin{split}
I_{max}
& = \rho ln \rho -\rho +I_0+S_0-\rho ln S_0 \\
& = I_0+(S_0-\rho)+\rho ln \frac{\rho}{S_0} \\
& =N-\rho+\rho ln \frac{\rho}{S_0} 
\end{split}
\end{equation}.

<br>

### Całkowita liczba osób zainfekowanych w trakcie epidemii

Z samych równań możemy również stwierdzić zmianę liczby zainfekowanych $I$. Przyjrzyjmy się równaniu $(5)$. Gdy $S_0>\rho$  liczba chorych będzie wzrastać, jednak liczba podatnych $S$ zawsze maleje, więc w pewnym momencie $S=\rho$, wtedy $I=I_{max}$, jednak $S$ ciągle maleje, więc później $S<\rho$, czyli liczba osób zainfekowanych $I$ zacznie spadać. Zauważmy, że będzie ona maleć do momentu osiągnięcia przez nią $0$. Jest to wartość krytyczna, i wtedy nie nastąpi już nigdzie żadna zmiana ilości, oznacza to, że wartości $S$ oraz $R$ pozostaną już na zawsze takie, jakie były gdy $I$ osiągnęło wartość $0$. Z wykresu fazowego (przedstawionego poniżej) można odczytać jaka będzie końcowa liczba osób podatnych po zakończeniu epidemii i są to wartości na osi $S$ różne w zależności od warunków początkowych. Omawiana liczba $S$ to liczba osób, które nie zostały zainfekowane podczas epidemii, oznacza to, że cała reszta przeszła chorobę. Nasuwa się więc pytanie, czy możemy wyznaczyć tą wartość. Okazuje się, że tak. Użyjemy wcześniej wykorzystanej sztuczki, czyli podzielimy dwa równania przez siebie $(4a)$ oraz $(4c)$:

$$\frac{dS}{dR}=-\frac{S}{\rho}.$$

Całkując otrzymujemy:

$$ln|S|=-\frac{1}{\rho}R+ln(C)$$

$$S=C e^{-\frac{R}{\rho}}.$$

Po obliczeniu stałej całkowania $C$ otrzymujemy

$$S(t)=S_0 e^{-\frac{R(t)}{\rho}}. \label{7}\tag{7} $$

Mamy więc zależność liczby podatnych $S$ od liczby wyzdrowiałych $R$. Interesuje nas jednak sytuacja dla konkretnego $t$, kóre jest czasem gdy $I$ przyjmie wartość $0$. Możemy więc zapisac, że interesuje nas $S(\infty)$, bo w nieskonczoności taka sytuacja na pewno nastąpi. Wtedy $R(\infty)=N-S(\infty)$ (bo $I(\infty)=0$), prowadzi nas to do wzoru

$$S(\infty)=S_0 exp\left(-\frac{R(\infty)}{\rho}\right)=S_0 exp\left(-\frac{N-S(\infty)}{\rho}\right),$$

czyli $S(\infty)$ jest dodatnim pierwiastkiem $z,$  $0<z<\rho,$ równania charakterystycznego

$$z=S_0 exp\left(-\frac{N-z}{\rho}\right). \label{8}\tag{8}$$

Teraz możemy podać całkowitą liczbę osób, które przeszły chorobę

$$I_{całkowite}=N-S(\infty),$$ 

gdzie $S(\infty)$ oznacza dodatnie rozwiązanie $z$ z równania $(8)$. Ta wartość jest oczywiście równa końcowej liczbie osób ozdrowiałych $R(\infty)$. Z powyższej analizy wyciągamy następujący ważny wniosek - ponieważ $I(t) \rightarrow 0$ i $S(t) \rightarrow S(\infty)$, więc epidemia wygasa z powodu braku osobników zainfekowanych, a nie z powodu braku podatnych $^{[6]}$.

<br>

### Tempo zdrowienia 

W przypadku większości epidemii pojawia się problem z wyznaczeniem liczby osób nowo zarażonych każdego dnia, ponieważ policzeni mogą zostać tylko Ci którzy poddali się leczeniu lub zdrowieją w inny sposób. Wobec tego, aby zastosować model do bieżącego problemu, potrzebujemy liczby zdrowiejących w jednostce czasu, czyli $dR/dt$, w funkcji czasu.

Z zależności $(7)$, równania $(4c)$  oraz tego że $S+I+R= N =const$ otrzymujemy równanie opisujące dynamikę $R$, gdzie $R(0) =0$:

$$
\begin{align}
\frac{dR}{dt} = \gamma I = \gamma (N-R-S) = \gamma (N-R-S_0 e^{-R/ \rho} )  \label{9}\tag{9},
\end{align}
$$

które można rozwiązać analitycznie tylko w sposoób parametryczny, a gdybyśmy znali $\beta$, $S_0$ oraz $N$ to łatwo rozwiązać to równanie numerycznie. Jednak zwykle nie znamy wszystkich parametrów, dlatego konieczne jest zastosowanie najlepszej procedury dopasowywania. Gdy epidemia jest mała, to znaczy $R/ \rho <1 $ możemy skorzystać z zaproponowanej przez Kermack'a I McKendrick'a w 1927 aprokrymacji równania postaci (w innych przypadkach do wyznaczenia $R(t)$ należy skorzystać z $(9)$) :

$$
\begin{align}
\frac{dR}{dt} = \gamma \left(N - S_0 + \left(\frac{S_0}{\rho } -1\right) R - \frac{S_0 R^2}{2 \rho ^2} \right) .
\end{align}
$$

Po rozkładzie prawej strony powyższego równania, scałkowaniu i po elementarnych, ale żmudnych obliczeniach otrzymujemy równanie $^{[5]}$:

$$
\begin{align}
\ R(t) = \frac{\beta ^2}{S_0} \left(\left( \frac{S_0}{\rho } -1 \right) + \alpha tgh \left(\frac{\alpha \gamma t}{2} - \phi \right)\right) ,
\end{align}
$$

gdzie:
- $ \alpha = \left( \left( \frac{S_0}{\rho} -1 \right)^2 + \frac{2 S_0 (N-S_0)}{\rho ^2} \right)^{1/2} $,

- $ \phi = \frac{tgh^{-1} \left(\frac{S_0}{\rho} -1 \right)}{\alpha}, $

z którego wynika, że <b> tempo zdrowienia </b> zadane jest jako: 

$$
\begin{align}
\ \frac{dR}{dt} = \frac{\gamma \alpha ^2 \rho ^2}{2 S_0} sech^2 \left(\frac{\alpha \gamma t}{2} - \phi \right).   \label{10}\tag{10}
\end{align}
$$

Zawiera ono tylko trzy parametry: $ \gamma \alpha^2 \rho^2 / (2 S_0), \alpha \gamma $ oraz $\phi$. 

<br>

#### Porównanie wyznaczonego tempa zdrowienia z danymi - bombajska epidemia dżumy (1905-1906) 

Na wykresie poniżej został przedstawiony model tempa zdrowienia z parametrami dopasowanymi przez Kermacka i McKendricka oraz rzeczywiste dane dotyczące bombajskiej epidemii dżumy $^{[9]}$. Ze względu na to, że w trakcie trwającej prawie rok epidemii większość ofiar zmarła, liczba osob wyzdrowiałych w ciągu tygodnia ($dR/dt$) została w tym przypadku zastąpiona liczbą ofiar śmiertelnych.

Równanie zaproponowane przez Kermacka i McKendricka miało postać $^{[5]}$:

$$
\begin{align}
\ \frac{dR}{dt} = 890 sech^2 (0,2 t - 3,4).  
\end{align}
$$

```python
import math

plt.figure(figsize=(8,5))
plt.title('Porównanie modelu z danymi (bombajska epidemia dżumy)')
plt.xlabel('czas (tygodnie)')
plt.ylabel('liczba zgonów')

#Model
def model(t):
    return 890*(1/(math.cosh(0.2*t -3.4)))**2

tygodnie = range(40)
liczba_zgonow = [model(T) for T in range(40)]
model = plt.plot(tygodnie, liczba_zgonow)

#Dane
dane = plt.plot(range(40), [16,8,8,12,16,24,48,51,92,124,178,280,387,442,644,
                           779,702,695,870,925,802,578,404,296,162,106,64,46,35,27,28,24,26,29,17,31,15,14,18,22], 'ro')
plt.show()
```

Po zakończeniu epidemi (gdy $ t\rightarrow \infty$) liczba zmarłych osobników wyniosła:

$$\ R(\infty) = \frac{\beta ^2}{S_0} \left(\left( \frac{S_0}{\rho } -1 \right) + \alpha \right), \hspace{15pt} \alpha = \left( \left( \frac{S_0}{\rho} -1 \right)^2 + \frac{2 S_0 (N-S_0)}{\rho ^2} \right)^{1/2}$$.

Różnica $N_0 - S_0$ jest równa początkowej liczbie osób zarażonych i jest ona znacznie mniejsza w porównaniu do $S_0$, można więc oszacować:
$$ R(\infty) \approx \frac{2 \beta ^2}{S_0}\left( \frac{S_0}{\rho } -1 \right). $$

## Model SIRS 

Model SIR zakłada, że każdy osobnik po wyzdrowieniu nabywa odporność na daną chorobę na całe życie (przestaje być podatny). W przypadku innej klasy chorób przenoszonych drogą powietrzną, na przykład grypy sezonowej, odporność danej osoby może z czasem maleć. Do opisu takich przypadków służy model SIRS (który umożliwia osobom wcześniej chorym powrotu do stanu podatnego).

<img src="SIRS.PNG"> 

Równania modelu SIRS są postaci: 

$$
\begin{align}
&\frac{dS}{dt} = - \beta S I + \xi R \\
&\frac{dI}{dt} = \beta S I - \gamma I  \\
&\frac{dR}{dt} = \gamma I - \xi R
\end{align}
$$

gdzie $ S$, $I$, $R$, $N$, $\beta$ oraz $\gamma$ oznaczają to samo co w modelu SIR, a $\xi$ - współczynnik powrotu do podatności na infekcję.

<br>

### Wizualizacja rozwoju epidemii dla modelu SIR oraz SIRS

Porównajmy modele SIR i SIR (wpływ współczynnika $\xi$ na przepływ osobników pomiędzy kolejnymi grupami) dla takich samych wartości początkowych $S_0$, $I_0$ oraz współczynników $\beta$, $\gamma$.

```python
def SS(x,s,i,r):
    return -beta*s*i + ksy*r                            
def IS(x,s,i,r): 
    return beta*s*i - gamma*i
def RS(x,s,i,r):
    return gamma*i - ksy*r

N_t = 3000
h=0.1
S_0, I_0, R_0 = 900, 100, 0

beta, gamma, ksy = 1/10000, 3/100, 0.7/1000

czas, S1, I1, R1 = rk4_method_3(S,I,R,S_0,I_0,R_0)
czas, S2, I2, R2 = rk4_method_3(SS,IS,RS,S_0,I_0,R_0)

fig, axes = plt.subplots(1,2, figsize=(22,7))
axes[0].plot(czas,S1,'g',czas,I1,'r',czas,R1,'b',linewidth=2)
axes[0].set_xlabel('czas (dni)',fontsize=15)
axes[0].set_ylabel('liczba osobników',fontsize=15)
axes[0].legend(['S','I','R'],loc="right")
axes[0].set_title('SIR',fontsize=20)

axes[1].plot(czas,S2,'g',czas,I2,'r',czas,R2,'b',linewidth=2)
axes[1].set_xlabel('czas (dni)',fontsize=15)
axes[1].set_ylabel('liczba osobników',fontsize=15)
axes[1].legend(['S','I','R'],loc="right")
axes[1].set_title('SIRS',fontsize=20)
plt.tight_layout()
plt.show()
```

# Model SEIR

## Model SEIR

W <b>SEIR</b>, czyli w kolejnym modelu epidemiologicznym, przepływy ludzi opisany jest pomiędzy czterema następującymi stanami: 
- S - osobniki zdrowe i podatne (ang.suscetible) na infekcję,
- E - osobniki narażone (ang. exposed),
- I - osobniki zainfekowane (ang. infected),
- R - osobniki wyleczone i odporne (ang. recovered). <br>

Nowy - osobniki narażone, to osobniki które po kontakcie z osobnikiem chorym (zainfekowanym) w stanie podatnym zostają zarażone, ale nie mogą jeszcze zarażać innych. Dopiero po przejściu do stanu chorych mogą zarażać innych podatnych osobników
Choroby, które opisuje ten model to na przykład ospa wietrzna, ptasia grypa (H7N7), ebola $^{[11]}$.

<img src="scheme/SEIR.PNG"> 

Równania modelu bazującego na założeniach analogicznych do modelu SI mają postać:

 jest to rozszerzenie modelu SIR o grupę zarażonych utajonych.
Osobniki po kontakcie z osobnikiem chorym (zainfekowanym) w stanie podatnym zostają zarażone, ale nie mogą jeszcze zarażać innych. Dopiero po przejściu do stanu chorych mogą zarażać innych podatnych osobników. Przejścia w tym modelu opisuje schemat

$$
\begin{align}
&\frac{dS}{dt} = - \beta S I  \\
&\frac{dE}{dt} = \beta S I - \sigma E   \\
&\frac{dI}{dt} = \sigma E - \gamma I \\
&\frac{dR}{dt} = \gamma I
\end{align}
$$

gdzie:
- $\beta$ - współczynnik (tempo) rozprzestzeniania się infekcji
- $\sigma$ - współczynnik ?? 
- $\gamma$ - współczynnik zdrowienia osobników zainfekowanych.

## Dynamika rozwoju chorób opisanych modelem SEIR

Poniżej na trzech wykresach zostanie przedstawiony rozwój epidemii dla wybranych chorób (ospa wietrzna, ptasia grypa oraz ebola) - zmiana liczby osobników w grupach $S,E,I,R$ od czasu.

Dane współczynniki są uśrednionymi wartościami pochodzącymi z różnych źródeł $^{[11]}$.

Zauważalne różnice pomiędzy wykresami dla ospy i ptasiej grypy a wykresem dla eboli wynikają ze sposobu rozprzestrzeniania się patogenów. Dla ospy i ptasiej grypy do zarażenia dochodzi również poprzez drogę powietrzną i kropelkową. Natomiast w przypadku eboli potrzebny jest bezpośredni kontakt z zainfekowanym osobnikiem.

Oś czasu na wykresach ograniczono do 30 tygodni, poniewa do tego czasu wszystkie epidemie wygasały.



```python
def S(x,s,e,i,r):
    return -beta*s*i   
def E(x,s,e,i,r):
    return beta*s*i - sigma*e                   
def I(x,s,e,i,r): 
    return sigma*e - gamma*i
def R(x,s,e,i,r):
    return gamma*i

S_0, E_0, I_0, R_0 = 999, 0, 1, 0

h = 0.1     
N_t = 210    # liczba dni 

fig, axes = plt.subplots(1,3, figsize=(22,7))

# OSPA WIETRZNA
beta = 0.001267 #/dzień
gamma = 0.133
sigma = 0.1
czas, S1, E1, I1, R1 = rk4_method_4(S,E,I,R,S_0,E_0,I_0,R_0)

axes[0].plot(czas,S1,'g',czas,E1,'y',czas,I1,'r',czas,R1,'b',linewidth=2)
axes[0].set_xlabel('czas (dni)',fontsize=15)
axes[0].set_ylabel('liczba osobników',fontsize=15)
axes[0].legend(['S','E','I','R'],loc="right")
axes[0].set_title('Ospa wietrzna \n' + r'$ \beta = $'+str(beta)+'   $\gamma = $'+str(gamma)+ '    $\sigma = $'+str(sigma),fontsize=20)

# PTASIA GRYPA
beta = 0.0004791
gamma = 0.083
sigma = 0.5
czas, S2, E2, I2, R2 = rk4_method_4(S,E,I,R,S_0,E_0,I_0,R_0)

axes[1].plot(czas,S2,'g',czas,E2,'y',czas,I2,'r',czas,R2,'b',linewidth=2)
axes[1].set_xlabel('czas (dni)',fontsize=15)
axes[1].set_ylabel('liczba osobników',fontsize=15)
axes[1].legend(['S','E','I','R'],loc="right")
axes[1].set_title('Ptasia grypa \n' + r'$ \beta = $'+str(beta)+'   $\gamma = $'+str(gamma)+ '    $\sigma = $'+str(sigma),fontsize=20)

# EBOLA
beta = 0.000382
gamma = 0.2857
sigma = 0.2985
czas, S3, E3, I3, R3 = rk4_method_4(S,E,I,R,S_0,E_0,I_0,R_0)

axes[2].plot(czas,S3,'g',czas,E3,'y',czas,I3,'r',czas,R3,'b',linewidth=2)
axes[2].set_xlabel('czas (dni)',fontsize=15)
axes[2].set_ylabel('liczba osobników',fontsize=15)
axes[2].legend(['S','E','I','R'],loc="right")
axes[2].set_title('Ebola \n' +  r'$\beta = $'+str(beta)+'   $\gamma = $'+str(gamma)+ '    $\sigma = $'+str(sigma),fontsize=20)

plt.tight_layout()
plt.show()
```
